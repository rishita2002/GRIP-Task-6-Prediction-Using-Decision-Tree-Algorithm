# Decision Tree - Train and Visualize
## Prediction using Decision Tree Algorithm
  ● Create the Decision Tree classifier and visualize it graphically.

  ● The purpose is if we feed any new data to this classifier, it would be able to predict the right class accordingly.

  Import Libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

## Converting into data frame

```python
iris = load_iris()
X = pd.DataFrame(iris.data[:,:], columns=iris.feature_names[:])
Y = pd.DataFrame(iris.target, columns=["Species"])
Y.sample(10)#setosa flowers are identified by zero, versicolor by one, and virginica by two.
```

## Setosa flowers are identified by 0, Versicolor by 1, and Virginica by 2.
   Applying Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier(max_depth=6) 
decision_tree.fit(X,Y)
print('Decision Tree Classifer Created')
```

## Visualizing the decision tree

```python
from sklearn import tree
fig = plt.figure(figsize=(40,50))
viz = tree.plot_tree(decision_tree, feature_names=X.columns, class_names=iris.target_names, filled=True)
```
